```python
questions = ["1. How many people have heart disease and how many people doesn't have heart disease?",
             "2. People of which sex has most heart disease?",
             "3. People of which sex has which type of chest pain most?",
             "4. People with which chest pain are most prone to have heart disease?",
             "5. People of which age has most heart disease?",
             "6. People of what age has most chest pain?",
             "7. Which cholestrol level has most heart disease?"]

questions
```

```python
#Let's find the answer of the first question.

#1. How many people have heart disease and how many people doesn't have heart disease?

#getting the values
df.target.value_counts()
```

```python
#plotting the bar chart.
df.target.value_counts().plot(kind = 'bar', color = ["orchid", "salmon"])
plt.title("Heart Disease values")
plt.xlabel("1 = Heart Disease, 0 = No heart Disease")
plt.ylabel("Amount");
```

```python
#plotting a pie chart.
df.target.value_counts().plot(kind = 'pie', figsize = (8, 6))
plt.legend(["Disease", "No disease"]);
```

```python
#SEX column part
#'0' represnt 'Female'
#'1' represent 'Male'

#Target column part
#'0' represnt 'No disease'
#'1' represent 'Disease'

#Now let's check how many 'male' and 'female' are in the dataset
df.sex.value_counts()
```

```python
#Plotting a pie chart.

df.sex.value_counts().plot(kind = 'pie', figsize = (8, 6))
plt.title('Male Female ratio')
plt.legend(['Male', 'Female']);
```

```python
#let's find the answer of the 2nd question.
#2. People of which sex has most heart disease?

pd.crosstab(df.target, df.sex)
```

```python
sns.countplot(x = 'target', data = df, hue = 'sex')
plt.title("Heart Disease Frequency for sex")
plt.xlabel("1 = Heart Disease, 0 = No heart Disease");
```

```python
#let's find the answer of the 3rd question.
#3. People of which sex has which type of chest pain most?

#counting values for diff chest pain
df.cp.value_counts()
```

```python
#plotting a bar chart
df.cp.value_counts().plot(kind = 'bar', color = ['salmon', 'lightskyblue', 'springgreen', 'khaki'])
plt.title('Chest pain type vs count');
```

```python
pd.crosstab(df.sex, df.cp)
```

```python
pd.crosstab(df.sex, df.cp).plot(kind = 'bar', color = ['coral', 'lightskyblue', 'plum', 'khaki'])
plt.title('Type of Chest pain for sex')
plt.xlabel('0 = Female, 1 = Male');
```

```python
#most of 'male' has 'type 0' chest pain at least of 'male' has 'type 4' pain.
#incase of 'female' 'type 0' and 'type 2' percentage is almost same.
```

```python
#let's find the answer of the 4th question.

#4. People with which chest pain are most prone to have heart disease?

pd.crosstab(df.cp, df.target)
```

```python
sns.countplot(x = 'cp', data = df, hue = 'target');
```

```python
#Most of the people have 'type 0' chest pain has less chance of heart disease.

#And we see the opposite for other types.

#Now let's take a look at our age column.

#create a distribution plot with normal distribution curve
sns.displot(x = 'age', data = df, bins = 30, kde = True);
```

```python
#'58-59' year old people are most in the dataset.

#Let's plot another distribution plot for 'maximum heart rate'
sns.displot(x = 'thalach', data = df, bins = 30, kde = True, color = 'chocolate');
```

```python
#let's find the answer of the 5th question.

#5. People of which age has most heart disease?
pd.crosstab(df.target, df.age)
```

```python
sns.countplot(x = 'target', data = df, hue = 'age')
plt.title('Heart disease frequency for age')
plt.xlabel('0 = No Heart disease, 1 = Heart disease')
plt.show()
```

```python
#let's find the answer of the 6th question.

#6. People of what age has most chest pain?
df.cp.value_counts()
```

```python
pd.crosstab(df.age, df.cp)
```

```python
pd.crosstab(df.age, df.cp).plot(kind = 'bar', color = ['coral', 'lightskyblue', 'plum', 'khaki'])
plt.title('Type of Chset Pain for Age')
plt.xlabel('0 = Female, 1 = Male');
plt.show()
```

```python
#let's find the answer of the 7th question.

#7. Which cholestrol level has most heart disease?
plt.figure(figsize = (7,4))
sns.histplot(x = df['chol'], kde = True, palette = 'rainbow', hue = df['target'])
plt.title('Cholestrol level having most heart disease')
plt.show()
```

```python
#Answer : Cholestrol level between 200 to 240 has most heart disease.
```
